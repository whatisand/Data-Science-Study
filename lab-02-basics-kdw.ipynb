### Tensorflow로 간단한 linear regression을 구현 (new)


```python
import tensorflow as tf


# x_train = [1, 2, 3]
# y_train = [1, 2, 3]


W_ = tf.Variable(tf.random_normal([1]), name='weight')
b_ = tf.Variable(tf.random_normal([1]), name='bias')
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])


hypothesis = X * W_ + b_

cost = tf.reduce_mean(tf.square(hypothesis - Y))

# tf.random_nomal 밖에 없는지
# tf.reduce_mean 이 무엇인지?

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(4002):
    
    cost_val, W_val, b_val, none_ = sess.run([cost, W_, b_, train], 
                                        feed_dict={X: [1,2,3,4,5],
                                                  Y: [2.1, 3.1, 4.1, 5.1, 6.1]})
    
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val, _)
        

        
```
